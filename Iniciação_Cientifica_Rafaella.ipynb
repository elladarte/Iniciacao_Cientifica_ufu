### Coletando os dados


Importando as Bibliotecas

```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt

import matplotlib
```

Criando a Função para realizar a coleta por meio da api do Banco Central. Para mais informações clique [aqui](https://dadosabertos.bcb.gov.br/dataset/10844-indice-de-precos-ao-consumidor-amplo-ipca---servicos/resource/c0980df7-ad92-47af-b71c-790825f4710a)

```python
def consulta_bc(codigo_bcb):
  url = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=json'.format(codigo_bcb)
  df = pd.read_json(url)
  df['data'] = pd.to_datetime(df['data'], dayfirst=True)
  df.set_index('data', inplace=True)
  return df
```

O codigo da serie historia do IPCA é 433, pode-se consultar atraves do site [Time Series Management System](https://www3.bcb.gov.br/sgspub/localizarseries/localizarSeries.do?method=prepararTelaLocalizarSeries), nela encontra-se a serie historica do IPCA desde de 02/01/1980 ate julho de 2021

```python
ipca = consulta_bc(433)
ipca
```

```python
ipca.to_csv("ipca.csv")
```

```python
ipca = consulta_bc(433)
start_data = '1999-01-01'
finish_data = '2021-01-01'
ipca_loc = ipca[ipca.index >= start_data]
ipca_loc = ipca_loc[ipca_loc.index <= finish_data]
```

```python
ipca_loc = ipca[ipca.index >= start_data]
ipca_loc = ipca_loc[ipca_loc.index <= finish_data]
ipca_loc
```

```python
fig = plt.figure(figsize=(18,12))
plt.plot(ipca_loc)
plt.title('Índice de Preços ao Consumidor Amplo entre os anos de 1999 a 2021')
plt.ylabel('Variação % por mês')
plt.xlabel('Anos')
```
